```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

sns.set()
sns.set_theme()
sns.set_style("white")
```

```python
pca_df = pd.read_csv("../results/Macaca_papio_pca.txt", sep="\t")
var_df = pd.read_csv("../results/Pca_var_explained.txt", sep="\t")
```

```python
pcx = 0
pcy = 1
chr_type = "Autosome"
reference_species = "Macaca_mulatta"


pca_df_macaca = pca_df.loc[(pca_df.REFERENCE == reference_species) & (pca_df.CHROM == chr_type)]
g = sns.scatterplot(pca_df_macaca, x="PC1", y="PC2", hue="SPECIES_SUBSPECIES")
g.set_title(chr_type)
plt.legend(bbox_to_anchor=(1.4, 1))
f = g.get_figure()
f.savefig("../figures/Macaca_aut_pca.jpg", dpi=300, bbox_inches='tight')
```
